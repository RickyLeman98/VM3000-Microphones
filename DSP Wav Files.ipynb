### Performing FFT on a real-world acquired test signal

```python
data, samplerate = soundfile.read("Audio Files/africa-toto.wav")
```

```python
print("Sampling rate is:", samplerate,"\n")
print("Wav file data array is:", data, "\n")

samples = data.shape[0]
print("Number of samples in wav file is:", samples,"\n")

print("Duration of wav file is:", data.shape[0]/samplerate, "seconds")
```

```python
plt.plot(data[:10*samplerate]) #Plotting first 10 seconds (stereo audio)
plt.title("Stereo audio wav file plot")
```

```python
data = data[:,0]
data.shape
```

```python
plt.plot(data[:10*samplerate]) #Plotting first 10 seconds (mono audio)
plt.title("Mono audio wav file plot")
```

```python
datafft = fft(data)
fftabs = abs(datafft)
freqs = fftfreq(samples,1/samplerate)
plt.xlim( [10, samplerate/2] )
plt.xscale( 'log' )
plt.ylim([0,20000])
plt.xlim([10,10000])
plt.grid( True )
plt.xlabel( 'Frequency (Hz)' )
plt.plot(freqs[:int(freqs.size/2)],fftabs[:int(freqs.size/2)])
plt.title("FFT output of mono audio from specified wav file")
```
